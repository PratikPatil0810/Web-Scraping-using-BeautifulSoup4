```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
for page in range(1, 4):
    url = "https://news.ycombinator.com/"
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    stories = soup.select("tr.athing")

    data = []

    for story in stories:
       
        rank = story.select_one(".rank")
        rank = rank.text.replace(".", "") if rank else "NA"

        title_tag = story.select_one(".titleline a")
        title = title_tag.text if title_tag else "NA"
        link = title_tag["href"] if title_tag else "NA"

        meta_row = story.find_next_sibling("tr")
        score_tag = meta_row.select_one(".score") if meta_row else None
        author_tag = meta_row.select_one(".hnuser") if meta_row else None
        comments_tag = meta_row.select_one("a:last-child") if meta_row else None

        points = score_tag.text if score_tag else "0 points"
        author = author_tag.text if author_tag else "NA"
        comments = comments_tag.text if comments_tag else "0 comments"

        data.append({
        "rank": rank,
        "title": title,
        "link": link,
        "points": points,
        "author": author,
        "comments": comments
        })

    df = pd.DataFrame(data)
    print(df.head())
```

```python
df.to_csv("hacker_news.csv", index=False)
```
